```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# (1, 90) (2,100) (3,50)
plt.plot([1,2,3],[90,100,50])
plt.show()
```

```python
plt.plot(["Math","Korean","English"],[90,100,50])
plt.plot(["Math","Korean","English"],[70,60,90])

plt.xlabel("Subjects")
plt.ylabel("Score")
plt.title("Midterm")

plt.legend(["Choi","Kim"])

plt.show()
```

```python
x = ["Math","Korean","English"]
y = [90,100,50]

plt.xlabel("Subjects")
plt.ylabel("Score")

plt.bar(x,y,width=0.2,color = "red")

plt.show()
```

```python
plt.plot([1,2,3],[90,100,50],marker = '+',markersize=20, linestyle="--")
plt.grid(True)
plt.axis([1,3,50,100])

plt.show()
```

# 판다스로 csv 데이터 불러오기

```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv")

train.head()
```

```python
import seaborn as sns
sns.set()
```

```python
sns.__version__
```

```python
sns.relplot(data=train,x= "Age",y="Survived")
```

```python
sns.relplot(data=train,x="Age",y="Survived",hue = "Sex",aspect=3)
```

```python
sns.catplot(data=train,x="Pclass",y="Fare",aspect=4,jitter = True,hue="Sex")
```

```python
sns.pairplot(data=train,hue="Sex",
            x_vars=["Pclass","Age","Fare"],
            y_vars=["Survived","Age","Fare"],height = 3)
```

```python
sns.catplot(data=train,x="Pclass",y="Fare",hue="Sex",kind="box")
```
