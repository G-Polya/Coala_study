```python
import pandas as pd

file_names = ["soviet.xlsx","공대생변승주.xlsx","난치.xlsx","미트.xlsx",
              "박병창.xlsx","슛뚜.xlsx","오늘의하늘.xlsx","오킹.xlsx",
             "온도.xlsx","윽박이.xlsx","이승인.xlsx","정선호.xlsx",
             "해그린달.xlsx","Park.xlsx"]
all_df = pd.DataFrame()
df = pd.read_excel(file_names[0])


for file in file_names:
    df = pd.read_excel(file)
    all_df = all_df.append(df,ignore_index = True)
all_df

all_df = all_df.drop(["Unnamed: 0","Unnamed: 0.1"], axis="columns")

all_df.to_excel("all_data.xlsx")
```
