```python
import pandas as pd

file_name = ["냔지","미트","박병창","이승인"]
for file in file_name:
    df = pd.read_excel(file+"6개월이내.xlsx")
    df = df.drop(["Unnamed: 0"],axis = "columns")
    
    썸네일 = df["썸네일타입"]
    제목 = df["title"]
    분야 = df["title_type"]
    조회수 = df["view_count"]
    좋아요수 = df["likes_count"]
    싫어요수 = df["dislikes_count"]
    한달업로드수 = df["term"]
    영상길이 = df["video_length"]
    
    df2 = pd.DataFrame({
        "썸네일" : 썸네일,
        "제목" : 제목,
        "분야" : 분야,
        "조회수" : 조회수,
        "좋아요 수" : 좋아요수,
        "싫어요 수":싫어요수,
        "한달업로드 수" : 한달업로드수,
        "영상길이":영상길이
    })
    
    df2.to_excel(file+"_최종규격.xlsx")
    print(df2)
    
```
