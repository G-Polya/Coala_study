```python
import pandas as pd
from datetime import datetime

file_name = ["soviet","난치","미트","공대생변승주","박병창","오늘의하늘",
            "오킹","온도","윽박이","이승인"]

all_df = pd.DataFrame()

for file in file_name:
    
    df = pd.read_excel(file+"6개월이내.xlsx")
    all_df = all_df.append(df,ignore_index = True)
    
all_df = all_df.drop(["Unnamed: 0.1.1","Unnamed: 0.1.1.1","Unnamed: 0","Unnamed: 0.1"], axis="columns")
all_df
```
